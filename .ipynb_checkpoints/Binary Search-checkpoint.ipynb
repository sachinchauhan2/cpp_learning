```c++
int BinarySearch(int arr[], int start,int end,int key)
{    
    int mid;
    while(start <= end)
    {
        mid = (start + end)/2;
        std::cout << "mid:"<< mid << std::endl;
        std::cout << arr[mid] << std::endl;
        std::cout <<"#########"<<std::endl;
        if (arr[mid] == key)
            std::cout << arr[mid] << std::endl;
            //return 1;
        
        if (arr[mid] < key)
            start = mid +1;
        //	std::cout<<start;
        else
            end = mid -1 ;
        
    }
    return 0;
}
```

```c++
int num[] = {10,23,45,70,90,100,111,124}
```
